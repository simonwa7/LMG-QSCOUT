```python
import matplotlib.pyplot as plt
# %pip install seaborn
import seaborn as sns
```

```python
ax = sns.scatterplot(range(1, 6), [0.014, 0.018, 0.043, 0.073, 0.070], label="optimized cry decomposition")
sns.scatterplot(range(1, 6), [0.039, 0.077, 0.145, 0.257, 0.321], ax=ax, label="naive cry decomposition")

ax.set_xlabel("Number of Qubits")
ax.set_ylabel("Absolute Relative Error")
ax.set_title("Error of LMG Benchmark for Different Circuit Decompositions")
plt.savefig("Error of LMG Benchmark for Different Circuit Decompositions.pdf", dpi=300)
```
